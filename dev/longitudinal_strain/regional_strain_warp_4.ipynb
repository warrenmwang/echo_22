What needs to happen now is to transform the respective frame's motion tracking information into a vector field that will allow us to properly warp these floating point coordinates to the next frame. What we have right now are floating point coordinates and integer point coordinates for the motion tracking so we can't apply the motion tracking to the points. We need to do some kind of transformation/interpolation on the motion tracking information to then warp the floating point coordinates. I have an idea to just loop over all of our floating point coordinates and find the four or more nearest integer point vectors in the motion tracking information and find an average vector direction and magnitude or something like that and apply that to the floating point coordinates.

```python
# let's just take the average of the 4 or 8 closest delta x and delta y values around the floating point 
coords_2 = []

for _ in coords_1:
    i,j = _[0], _[1]
    
    # get an average displacement for this point from the nearest integer displacements
    # first round down
    i_int = int(i)
    j_int = int(j)
    
    reference_i_js = [ [i_int, j_int],
                       [i_int - 1, j_int - 0],
                       [i_int - 0, j_int - 1],
                       [i_int - 1, j_int - 1],
                       [i_int + 1, j_int + 0],
                       [i_int + 0, j_int + 1],
                       [i_int + 1, j_int + 1],
                       [i_int + 1, j_int - 1],
                       [i_int - 1, j_int + 1] ]
    
    delta_i = 0.0
    delta_j = 0.0
    count = 0 
    
    for pair in reference_i_js:
        i_tmp, j_tmp = pair[0], pair[1]
    
        delta_j_tmp = curr_clip_motions[0][1][i_tmp][j_tmp] # forward x = forward j
        delta_i_tmp = curr_clip_motions[1][1][i_tmp][j_tmp] # forward y = forward i
        
        delta_i += delta_i_tmp
        delta_j += delta_j_tmp
    
        count += 1
    
    i_new = i + (delta_i / count)
    j_new = j + (delta_j / count)
    
    coords_2.append([i_new, j_new])
    
coords_2 = np.array(coords_2)
```

```python
def foo(coords, delta_ed_es):
    '''
    coords - (N, 2) ndarray not tensor
    delta_ed_es - int
    '''
    coords_return = None
    in_coords_len = coords.shape[0]
    
    for frame in range(delta_ed_es):
        # first warp from ED -> ED+1
        if frame == 0:
            # warp forward all the current points
            coords_1 = []
            for _ in range(len(coords)):
                pair = coords[_]

                i_0,j_0 = pair[0],pair[1]

                # forward change in y
                delta_i = curr_clip_motions[1][frame][i_0][j_0]

                # forward change in x
                delta_j = curr_clip_motions[0][frame][i_0][j_0]

                i_1, j_1 = i_0 + delta_i , j_0 + delta_j

                coords_1.append([i_1, j_1])
            # for second iteration
            coords_return = coords_1
        else:
            # all other warps starting from ED+1 -> ED+2
            # to final warp ED+(n-1) -> ED+n
            coords_2 = []

            for _ in coords_return:
                i,j = _[0], _[1]

                # get an average displacement for this point from the nearest integer displacements
                # first round down
                i_int = int(i)
                j_int = int(j)

                reference_i_js = [ [i_int, j_int],
                                   [i_int - 1, j_int - 0],
                                   [i_int - 0, j_int - 1],
                                   [i_int - 1, j_int - 1],
                                   [i_int + 1, j_int + 0],
                                   [i_int + 0, j_int + 1],
                                   [i_int + 1, j_int + 1],
                                   [i_int + 1, j_int - 1],
                                   [i_int - 1, j_int + 1] ]

                delta_i = 0.0
                delta_j = 0.0
                count = 0 

                for pair in reference_i_js:
                    i_tmp, j_tmp = pair[0], pair[1]

                    delta_j_tmp = curr_clip_motions[0][frame][i_tmp][j_tmp] # forward x = forward j
                    delta_i_tmp = curr_clip_motions[1][frame][i_tmp][j_tmp] # forward y = forward i

                    delta_i += delta_i_tmp
                    delta_j += delta_j_tmp

                    count += 1

                i_new = i + (delta_i / count)
                j_new = j + (delta_j / count)

                coords_2.append([i_new, j_new])
                
            # prep for next iter
            coords_return = coords_2
            
    return np.array(coords_return)
```

```python
# now im going to guess a way to calculate a new strain because our old way of comparing the number
# of white pixels in each frame won't work anymore now that we are looking at points instead of 
# pixels in an image

# for each point in ES, calculate the distance moved relative to the same point in the ED frame
def get_distance(x2,y2,x1,y1):
    return np.sqrt((x2-x1)**2 + (y2-y1)**2)

total_diff = 0
for i_0, j_0, i_1, j_1 in zip(coords_i_vals, coords_j_vals, coords_ES_i_vals, coords_ES_j_vals):
    total_diff += get_distance(j_1,i_1,j_0,i_0)
```
